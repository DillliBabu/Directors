# Director Dataset Project

```python
import pandas as pd
import numpy as np
```

```python
url='https://raw.githubusercontent.com/DillliBabu/Directors/refs/heads/main/directors.csv'
df=pd.read_csv(url)
```

```python
df.rename(columns={'director_name':'Director_Name','id':"ID","gender":"Gender"},inplace=True)
df.head()
```

```python
# Rename the column director_name to director.
df.rename(columns={'Director_Name':'Director'},inplace=True)
df.head()
```

```python
# Drop the column Unnamed: 0.
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
# Check if there are any null values in the dataset.
df.isnull().sum()
```

```python
# Count total number of directors in the dataset.
df['Director'].count()
```

```python
# Find all directors whose gender is Male.
df[df['Gender']=='Male']
```

```python
# Find all directors whose gender is Female.
df[df['Gender']=='Female']
```

```python
# Count how many directors are Male and Female.
df['Gender'].value_counts()
```

```python
# Display only director_name and gender columns.
df[['Director','Gender']]
```

```python
# Find the number of unique directors.
df['Director'].unique()
```

```python
# Check if any duplicate director names exist.
df['Director'].duplicated().sum()
```

```python
# Remove duplicate director records if any.
df['Director'].drop_duplicates(inplace=True)
```

```python
# Sort the DataFrame by director_name in ascending order.
df.sort_values(by='Director',ascending=True)
```

```python
# Sort the DataFrame by id in descending order.
df.sort_values(by='ID', ascending=False)
```

```python
# Find the director with the minimum id.
df.loc[df['ID'].idxmin(),'Director']
```

```python
# Find the director with the maximum id.
df.loc[df['ID'].idxmax(),'Director']
```
